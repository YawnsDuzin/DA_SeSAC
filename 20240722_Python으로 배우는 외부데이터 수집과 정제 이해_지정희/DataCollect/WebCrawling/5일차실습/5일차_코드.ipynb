## (실습) Selenium / 네이버 쇼핑

````python
def scroll_down(driver):
    # ================================================================
    # 전체 스크롤 처리
    prods = driver.find_elements(By.CSS_SELECTOR, 'div.basicList_list_basis__uNBZx > div > div')
    print('스크롤 전 - len(prods) :', len(prods))

    last_height = driver.execute_script('return document.body.scrollHeight')
    print('last_height : ', last_height)

    while True:
        # 웹 브라우저 스크롤
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
        time.sleep(random.uniform(0.5, 1))      # 딜레이(0.5 ~ 1 sec)

        current_height = driver.execute_script("return document.body.scrollHeight")
        print('last_height : ', last_height, ', current_height : ', current_height)

        # 더 이상 스크롤 할게 없다!
        if last_height == current_height:
            break

        last_height = current_height
    # ================================================================
````

## 새로운 페이지 접근하기
- 새로운 탭에 웹페이지가 오픈되는 경우
    - Diner 객체는 하나의 웹 페이지 만 접근할 수 있음
    - 새로운 페이지로 접근하기 위해서는 driver.switch_to.window() 함수 이용
    - 사용방법
        - 현재 윈도우 핸들 확인하기
        ```
        driver.current_window_handle
        ```
        - 모든 윈도우 핸들 확인하기
        ```
        driver.window_handles
        ```
        - 윈도우 전환하기
        ```
        driver.switch_to.window(driver.window_handles[1])
        ```

````python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

import time
import random
import pandas as pd

# 크롬 드라이버 생성
driver = webdriver.Chrome()

# 웹 페이지 열기 (네이버 메인)
url = 'https://www.naver.com/'
driver.get(url)

# 화면 최대화
driver.maximize_window()

# 검색창 찾기 (검색창 클릭 -> 검색어입력('복숭아') -> 조회(엔터키입력))
search = driver.find_element(By.CSS_SELECTOR, '#query')
search.click()
search.send_keys('딸기', Keys.ENTER)      # 키 같이 보내기


#main_pack > section.sc_new.sp_nshop._shopping_root > div:nth-child(1) > div.mod_more_wrap._more_root
# 쇼핑더보기 찾기
search = driver.find_element(By.CSS_SELECTOR, 'div.mod_more_wrap._more_root')
search.click()

print('화면 전환 전 : ', driver.title, ', current_window_handle : ', driver.current_window_handle)

# 새로 생긴 탭으로 전환
lst_wh = driver.window_handles
for wh in lst_wh:
    print(wh)
driver.switch_to.window(lst_wh[1])

print('화면 전환 후 : ', driver.title, ', current_window_handle : ', driver.current_window_handle)

lst = []

max_page = 3
for i in range(1, max_page + 1):
    # 전체 스크롤
    scroll_down(driver)

    prods = driver.find_elements(By.CSS_SELECTOR, '.product_item__MDtDF')         # 광고제품 제외
    print('Page : ', i, ', 스크롤 후 - len(prods) :', len(prods))

    for prod in prods:
        name = prod.find_element(By.CSS_SELECTOR, '.product_title__Mmw2K > a').text.strip()
        link = prod.find_element(By.CSS_SELECTOR, '.product_title__Mmw2K > a').get_attribute('href')
        price = prod.find_element(By.CSS_SELECTOR, 'div.product_price_area__eTg7I > strong > span.price > span > em').text.strip().replace(',', '')

        lst.append([name, link, price])

    print(lst)

    #content > div.style_content__xWg5l > div.pagination_pagination__fsf34 > a
    if i < max_page:
        next = driver.find_element(By.CSS_SELECTOR, '.pagination_num__b1BF2 + a')
        next.click()
        time.sleep(random.uniform(1,2))
````

## IFrame 제어

- IFrame 이란?
    - HTML 문서 내에 다른 HTML 문서를 삽입할 수 있는 태그
        - 한 웹 페이지 안에 다른 웹 페이지를 포함시킬 때 사용
    - IFrame 안의 콘텐츠는 메인 페이지와는 독립된 브라우저 환경을 가짐
        - 메인 페이지와는 별개의 자바스크립트 실행, 스타일 적용이 수행 됨
- 사용 예
    - 네이버 블로그 또는 카페
        - 외부 컨텐츠를 포스트에 포함시킬때 IFrame 사용
            - 유튜브 동영상이나 지도를 포함시키는 경우
            - 광고 배너 삽입
        - 외부에서 제공하는 양식 작성
            - 글 작성할때 웹 페이지를 떠나지 않고도 양식을 작성할 수 있음
        - ***동적 컨텐츠를 비동기 방식으로 로딩***
            - 특정 컨텐츠를 비동기로 로딩하여 페이지 로드 시간을 단축하거나, 메인 페이지의 성능을 영향을 주지 않음

- 셀리니움에서 iframe 제어하는 방법
    - 웹 페이지를 제어할 때, iframe 안의 요소를 접근하기 위해서는 iframe 전환해야 함.
    - driver.switch_to.frame() 함수를 이용하여 iframe 으로 전환
        - 이름 또는 아이디로 전환하기
            ```
            driver.switch_to_frame("iframe_id)
            ```
        - 인덱스로 전환하기
            ```
            driver.switch_to_frame(0)
            ```
        - WebElement 사용하기
            ```
            iframe_element = driver.find_element_by_tag_name('iframe')
            driver.switch_to.frame(iframe_element)
            ```
    - iframe에서 벗어나기
        - 기본 콘텐츠로 돌아가기
            ```
            driver.switch_to.default_content()
            ```

````python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

import time
import random
import pandas as pd

# 크롬 드라이버 생성
driver = webdriver.Chrome()

# 네이버 쇼핑몰 열기
url = 'https://search.shopping.naver.com/search/all?adQuery=%EB%94%B8%EA%B8%B0&frm=NVSCPRO&origQuery=%EB%94%B8%EA%B8%B0&pagingIndex=1&pagingSize=40&productSet=total&query=%EB%94%B8%EA%B8%B0&sort=rel&timestamp=&viewType=list'
driver.get(url)

# 화면 최대화
driver.maximize_window()

# 전체 스크롤
scroll_down(driver)

print(driver.title)

# iframe 변경
driver.switch_to.frame("powerlink-div_tgtLREC")

# .power_list > li
links = driver.find_elements(By.CSS_SELECTOR, '.power_item')
#gfp_sf_align > div > div > ul > li:nth-child(1) > div > div.summary > p
lst = []
for link in links:
    name = link.find_element(By.CSS_SELECTOR, '.power_title.pcol3').text
    des = link.find_element(By.CSS_SELECTOR, 'div.summary > p').text
    link = link.find_element(By.CSS_SELECTOR, '.power_url').text

    lst.append([name, des, link])

print(lst)

# iframe 나가기 (이전 iframe 으로)
driver.switch_to.default_content()
print(driver.title)
````

````python
import pandas as pd
df = pd.DataFrame(lst, columns = ['제품명', '설명', '링크'])
df
````
