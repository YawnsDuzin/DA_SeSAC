## 실습 (네비어 웹툰 도전 만화 스크래핑하기)
- url = https://comic.naver.com/challenge
- 타이틀, 링크, 별점, 조회수
- 제출물
    - .ipynb 파일과 생성된 엑셀파일

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time
import random
import pandas as pd

# 크롬 드라이버 생성
driver = webdriver.Chrome()

lst = []
i = 0
while(True):
    i += 1

    # naver웹툰 -> 도전만화
    url = f'https://comic.naver.com/challenge?page={i}'
    driver.get(url)

    time.sleep(random.uniform(1,3))
    WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'a.ContentTitle__title_area--x24vt > span > span')))
    
    links = driver.find_elements(By.CSS_SELECTOR, '.ChallengeView__challenge_list--vEhFq > li.ChallengeListItem__item--X3Md7')
    
    if links == None: break
    
    print(url)
    
    #content > div:nth-child(2) > ul > li:nth-child(1)
    #content > div:nth-child(2) > ul > li:nth-child(1) > div > a.ContentTitle__title_area--x24vt > span > span
    
    for link in links:
        # 타이틀, 링크, 별점, 조회수
        #content > div:nth-child(2) > ul > li:nth-child(1) > div > a.ContentTitle__title_area--x24vt > span > span
        title = link.find_element(By.CSS_SELECTOR, 'a.ContentTitle__title_area--x24vt > span > span').text
        link_2 = link.find_element(By.CSS_SELECTOR, '.ChallengeListItem__info_area--hnJTz > a:nth-child(1)').get_attribute('href')
        rank = link.find_element(By.CSS_SELECTOR, 'a.ContentTitle__title_area--x24vt > span > span').text
        
        writer = link.find_element(By.CSS_SELECTOR, '.ContentAuthor__author--CTAAP').text
        cnt = link.find_element(By.CSS_SELECTOR, '.Rating__view_area--GQb_S > span').text
        page = i
        
        lst.append([title, link_2, rank, cnt, writer, page])
    
    # 0     : Page 8
    # 1 ~ 2 : Page 28
    # 5 ~ 10 : 
    #time.sleep(random.uniform(5,10))
    
print(lst)
```

```python
df = pd.DataFrame(lst, columns = ['타이틀', '링크', '별점', '조회수', '작가', 'page'])
df
```

```python
df.to_excel('naver_webtoon.xlsx', index = True)
```
