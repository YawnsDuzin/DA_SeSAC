# json 데이터 읽고 쓰기

## json.loads()
- json 문자열을 파이썬 사전 객체로 디코딩

```python
# json.loads() : json 문자열을 파이썬 사전 객체로 디코딩
result = json.loads(obj)
print(type(result))         # <class 'dict'>
result
```

```python
print('name : ', result['name'])
print('place_lived : ', result['place_lived'])
print('pet : ', result['pet'])
print('siblings : ', result['siblings'])
```

```python
print(result['siblings'][1]['name'])
print(result['siblings'][1]['age'])
print(result['siblings'][1]['pet'])
```

## json.load()
- 파일에서 읽은 JSON데이터를 파이썬 사전 객체로 디코딩

```python
# json.load() : 파일에서 읽은 JSON데이터를 파이썬 사전 객체로 디코딩
json_file_path = 'sample.json'
with open(json_file_path, 'r', encoding='utf-8') as file:
    result1 = json.load(file)

result1
```

```python
print(result1['siblings'][0]['name'])
print(result1['siblings'][0]['age'])
print(result1['siblings'][0]['pet'])
```

```python
result1['siblings'][1]['age'] = 30    # 나이를 30으로 변경

print(result1['siblings'])
```

## json.dumps()
- 파이썬 사전 객체를 JSON 문자열로 인코딩

```python
new_result = json.dumps(result1)
print(new_result)
print(type(new_result))
```

```python
# json.dumps() : 파이썬 사전 객체를 JSON 문자열로 인코딩

# a python object (dict)
x = {
    "name" : "John",
    "age" : 30,
    "city" : "New York"
}

y = json.dumps(x)

print('type(x) : ', type(x))
print('type(y) : ', type(y))
print(y)
```

```python
# Python 객체를 JSON 문자열로 변환하고 값을 출력합니다.
import json

print(json.dumps({"name" : "John", "Age" : 30}))
print(json.dumps(['apple', 'banana']))
print(json.dumps(('apple', 'banana')))
print(json.dumps('hello'))
print(json.dumps(42))
print(json.dumps(31.76))
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

```python
import json
x = {
    "name" : "John",
    "age" : 30,
    "married" : True,
    "divorced" : False,
    "children" : ("Ann", "Billy"),
    "pets" : None,
    "cars" : [
        {"model" : "BMW 230", "mpg" : 27.5},
        {"model" : "Ford Edge", "mpg" : 24.1}
    ]
}

print(json.dumps(x))
```
