```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import openpyxl
```

### 네이버 지식인에서 키워드 검색 결과 파싱하기
제목 / 링크 / 날짜 / 카테고리 / 답변수

```python
header = {
    'User-Agent' : 'Mozilla/5.0',
    'Referer' : 'https://kin.naver.com/search/list.naver?query=%EB%B9%85%EB%8D%B0%EC%9D%B4%ED%84%B0'
}

lst = []

for i in range(1,10):
    url = f'https://kin.naver.com/search/list.naver?query=%EB%B9%85%EB%8D%B0%EC%9D%B4%ED%84%B0&page={i}'
    response = requests.get(url, headers=header)

    if response.status_code == 200:
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')

        items = soup.select('.basic1>li')

        for item in items:
            title = item.select_one('.\\_nclicks\\:kin\\.txt').text.strip()
            link = item.select_one('.\\_nclicks\\:kin\\.txt').attrs['href'].strip()
            date = item.select_one('.txt_inline').text.strip()
            category = item.select_one('.txt\\_g1.\\_nclicks\\:kin\\.cat2').text.strip()
            answers = item.select_one('.hit').text.strip()
            lst.append([title, link, date, category, answers])

    df = pd.DataFrame(lst, columns=['제목','링크','날짜','카테고리','답변수'])
    df.to_excel('naver.xlsx')


    # else:
    #     print("실패 : ", response.status_code)





```
