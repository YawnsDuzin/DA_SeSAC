```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

url = "https://kin.naver.com/search/list.naver?query=%EB%B9%85%EB%8D%B0%EC%9D%B4%ED%84%B0"
response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    QandA = soup.select('dl')
    data = []
    for page in QandA:
        title = page.select_one('dt').text.strip()
        link = page.select_one('dt>a').attrs['href'].strip()
        date = page.select_one('.txt_inline').text.strip()
        catg = page.select_one('.txt\\_g1.\\_nclicks\\:kin\\.cat2').text.strip()
        answ = page.select_one('.hit').text.strip()
        data.append([title, link, date, catg, answ])
    df = pd.DataFrame(data, columns=['제목','링크','날짜','카테고리','답변'])
    # print("?")
    df.to_excel('naverQandA.xlsx',index=False)
else:
    print('실패 : ',response.status_code)
df
```
