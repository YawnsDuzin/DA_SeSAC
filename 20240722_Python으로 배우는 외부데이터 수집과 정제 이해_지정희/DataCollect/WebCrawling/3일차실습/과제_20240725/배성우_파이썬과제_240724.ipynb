```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

Q = input("검색할 단어를 입력하시오:")
count = int(input("몇페이지까지 가져올건가요? : "))+1


df = {
    "제목" : [],
    "링크" : [],
    "날짜" : [],
    "카테고리" : []

}

for i in range(1,count):
    url = f"https://kin.naver.com/search/list.naver?query={Q}&page={i}"
    response = requests.get(url)
    if response.status_code == 200:
        html = response.text
        soup = BeautifulSoup(html,'lxml')
        items = soup.select_one(".basic1")

        for item in items.find_all('li'):
          ㄹ title = item.select_one("dt a").text.strip()
           link = item.select_one("dt a").attrs['href'].strip()
           date = item.select_one(".basic1 dl .txt_inline").text.strip()
           category = item.select_one(".txt_block a + a").text.strip()

           df['제목'].append(title)
           df['링크'].append(link)
           df['날짜'].append(date)
           df['카테고리'].append(category)

    else:
        print("ㅠㅠ")

df= pd.DataFrame(df)
df
```
