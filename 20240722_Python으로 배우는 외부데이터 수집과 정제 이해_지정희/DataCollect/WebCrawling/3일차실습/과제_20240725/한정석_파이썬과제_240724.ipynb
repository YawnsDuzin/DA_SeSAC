# 네이버 지식인 키워드검색  
### 제목, 링크, 날짜, 카테고리, 답변수

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import openpyxl
import re


url_main = 'https://kin.naver.com/'

query = input('검색할거:')
page = input('검색할 페이지수:')

data = []
for i in range(1,int(page)+1):
    url = 'https://kin.naver.com/search/list.naver'
    param = {
        'query': query,
        'page': i
    }
    response = requests.get(url=url, params=param)
    if response.status_code == 200:
        html = response.content
        soup = BeautifulSoup(html, 'html.parser')

        items = soup.select('#s_content > div.section > ul > li')


        for item in items:
            #제목
            title = item.select_one('dl > dt > a').text.strip()
            #링크
            link = item.select_one('dl > dt > a').attrs['href']
            #날짜 
            year = item.select_one('dl > dd.txt_inline').text.strip()
            #카테고리 
            category = item.select_one('dd > a.txt_g1._nclicks\\:kin\\.cat2').text.strip()
            #답변수
            ans_count = item.select_one('dd.txt_block > span.hit').text.strip()
            acn = int(re.findall(r'\d+', ans_count)[0])
            
            data.append([title,link,year,category,acn])
            df = pd.DataFrame(data=data, columns=['제목','링크','날짜','카테고리','답변수'])
        if int(page) == (int(page)+1):
            break
    else:
        print('request 실패',response.status_code)
df.to_excel('지식인.xlsx', index=False)
```
