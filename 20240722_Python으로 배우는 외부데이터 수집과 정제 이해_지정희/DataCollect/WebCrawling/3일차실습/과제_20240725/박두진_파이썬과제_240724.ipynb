## 과제 사용법 실습하기
- 네이버 지식인에서 키워드 검색 (예. 빅데이터) 결과 파싱하기
    - https://kin.naver.com
    - https://kin.naver.com/search/list.naver?query=%EC%98%A4%ED%94%88ai&page=2
    - 제목,링크,날짜,카테고리,답변수
        - 주의) 클래스 이름에 특수문자가 있는 경우 \\처리 해야 함.
        - 사용자가 정한 페이지 수 만큼 가져오기

```python
import requests
import pandas as pd
from urllib.parse import quote

def kin_naver_parse():#(query = '', page = 1):    
    query = input('조회 할 검색어를 입력해주세요 =>')
    page = input('조회 할 페이지 수를 입력해주세요 =>')
    
    lst = []
    i = 0
    
    print('query-1 :' , query)
    query = quote(query, safe='')       # 한글 인코딩
    print('query-2 :' , query)
    
    while(True):
        i += 1        
        
        url = f'https://kin.naver.com/search/list.naver?query={query}&page={i}'
        header = {
            'User-Agent' : 'Mozilla/5.0',
            'Referer' : url
        }

        print(url)
        
        response = requests.get(url, headers=header)
        result = response.text

        from bs4 import BeautifulSoup

        soup = BeautifulSoup(result, 'html.parser')
        #print(soup.prettify())

        items = soup.select("ul.basic1 > li")
        # print(type(items))
        # print(len(items),'\r\n', items)
    
        for news in items:
            # 제목,링크,날짜,카테고리,답변수
            제목 = news.select_one('dl > dt > a').text.strip()
            링크 = news.select_one('dl > dt > a').attrs['href'].strip()
            날짜 = news.select_one('dl > dd.txt_inline').text.strip()
            카테고리 = news.select_one('dl > dd.txt_block > a.txt_g1._nclicks\\:kin\\.cat2').text.strip()
            답변수 = news.select_one('dl > dd.txt_block > span.hit').text.strip().replace('답변수', '')
          
            lst.append([제목, 링크, 날짜, 카테고리, 답변수])
        
        print(lst)
        if page == str(i): break

    df = pd.DataFrame(lst, columns = ['제목', '링크', '날짜', '카테고리', '답변수'])
    return df
```

```python
df = kin_naver_parse()
df
```
