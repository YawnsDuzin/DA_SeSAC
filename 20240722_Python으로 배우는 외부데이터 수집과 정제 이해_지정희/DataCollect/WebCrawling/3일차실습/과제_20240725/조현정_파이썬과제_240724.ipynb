   - class = "kin.txt" 이면, class = "kin\\.txt"로 표시해줘야 함.  
    ('.'이 클래스 속성을 나타내는 게 아니라 특수문자 표시이다.)

```python
## 네이버 지식인에서 검색할 키워드를 사용자에게 입력 받음
## 사용자가 원하는 페이지 수만큼 페이지를 스크래핑
## 제목/링크/날짜/카테고리/답변수 테이블로 엑셀파일 생성
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

## 네이버 지식인에서 검색할 키워드를 사용자에게 입력 받음
keyword = input('검색할 키워드를 입력하세요 : ')

# 사용자가 원하는 페이지수만큼 페이지를 스크래핑
pages = int(input('가져오고 싶은 페이지 수를 입력하세요 : '))

params = {
    'query': keyword,
    'page': pages}


data = [] # for문을 한 번 돌 때마다 추가된 data 리스트가 초기화되지 않도록 전체 반복문 밖에서 초기값을 지정해야 함.
for page in range(1,pages+1):
    url = f'https://kin.naver.com/search/list.naver' # 1-1
    headers = {'User-Agent':'Mozilla/5.0', 'Referer': url} # 1-2 위장할 User-Agent, Referer 설정
    response = requests.get(url, headers = headers, params=params) # 2
    root_domain = 'kin.naver.com' #### 스크래핑한 링크가 xlsx파일에서 바로 열릴 수 있게 도메인 주소 추가 ####

    if response.status_code == 200:  # 3
        result = response.text   # 4
        soup = BeautifulSoup(result, "html.parser") #5 html을 파싱한 soup객체 생성
        search = soup.select('#s_content > div.section > ul > li')

        for item in search:
            title = item.select_one('dl > dt > a').text.strip()
            link = root_domain + item.select_one('dl > dt > a').attrs['href']
            date = item.select_one('dd.txt_inline').text.strip()
            category = item.select_one(r'dl > dd.txt_block > a.txt_g1._nclicks\:kin\.cat2').text.strip()
            replies = item.select_one('dl > dd.txt_block > span.hit').text.strip()
            replies_num = int(re.findall(r'\d+',replies)[0])

            data.append([title,link,date,category,replies_num])

        df = pd.DataFrame(data, columns=['제목','링크','날짜','카테고리','답변수'])
        df.to_excel('naver_QnA_Seaching_result.xlsx', index=False)


        if page == (pages+1):
            break




    else:
        print("실패 : ",response.status_code)

df
```
