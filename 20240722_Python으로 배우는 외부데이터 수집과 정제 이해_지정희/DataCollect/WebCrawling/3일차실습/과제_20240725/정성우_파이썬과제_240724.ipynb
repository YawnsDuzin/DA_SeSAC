```python
from bs4 import BeautifulSoup
import requests, pandas as pd
 
keyword = input('검색 키워드를 입력하세요 :')
pages = int(input('검색할 페이지 수를 입력하세요 :'))
header = {
    'User-Agent' : 'Mozilla/5.0',
    'Referer' : 'https://kin.naver.com/search/list.naver'
}
data = []
for i in range(1, pages + 1):
    url = 'https://kin.naver.com/search/list.naver'
    param = {
        'query': keyword,
        'page': i
    }
    response = requests.get(url, headers=header, params=param)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        items = soup.select('.basic1>li')
        for item in items:
            data.append([item.select_one('dl a').text, item.select_one('dl a')['href'], item.select_one('dl .txt_inline').text, item.select_one('dl .txt_block a:nth-of-type(2)').text, item.select_one('dl .hit').text])
    else:
        print("실패 :", response.status_code)

df = pd.DataFrame(data, columns=['질문제목', '링크', '날짜', '카테고리', '답변수'])
df.to_excel('naverkin.xlsx', index=False)
df
```
