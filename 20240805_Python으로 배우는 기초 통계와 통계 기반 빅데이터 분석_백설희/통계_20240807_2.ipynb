### 1. 넘파이의 순열 함수를 이용하여 로또번호 10세트를 출력하시오
- 순열(비복원, 순서정렬)
- from numpy.random import permutation

```python
permutation(45) + 1
```

```python
import numpy as np
from numpy.random import permutation

arr1 = np.arange(1, 46)
print('arr1 : ', arr1)

lst_lotto = []

# ============================================
# 6번 shuffle
# for i in range(10):
#     lst_lotto.append(permutation(arr1)[:6])

# ============================================
# 1번 shuffle
i_idx = 0
pmt = permutation(arr1)
print('pmt : ', pmt)

for i in range(10):
    lst_lotto.append(pmt[i_idx:i_idx + 6])
    i_idx += 6

print(lst_lotto)
```

### 2. x1, x2를 산점도로 시각화하라
- x1은 평균이 5.0이고 표준편차가 2.0인 정규분포에서 임의 추출한 100개의 배열이고,
- x2는 평균이 3.0이고 표준편차가 1.0인 정규분포에서 임의 추출한 100개의 배열이다.

```python
from numpy.random import normal
```

```python
import seaborn as sns

N = 10000
x1 = np.random.normal(loc = 5, scale = 2, size = N)
x2 = np.random.normal(loc = 3, scale = 1, size = N)

# sns.scatterplot(x = x1, y = x2)
# sns.relplot(x = x1, y = x2, kind = 'scatter')


# 2차원 - 산점도 + 히스토그램
# sns.jointplot(x = x1, y = x2, space=0.2)

sns.jointplot(x = x1, y = x2,
              kind = 'reg',      # 선형회귀선, KDE 밀도곡선 추가
              space = 0,          # x축, y축 공간 0으로 만들기
              #size = 5,           # 크기 조정하기
              ratio = 4           # 비율 조정하기
              )
```

```python
print(x1.mean(), x1.std())
print(x2.mean(), x2.std())
```

#### 균등분포

```python
from numpy.random import uniform

N = 10000
x1 = uniform(low=100.0, high = 200.0, size = N)
x2 = uniform(low=70.0, high = 0.0, size = N)

sns.jointplot(x = x1, y = x2,
              kind = 'reg',       # 선형회귀선, KDE 밀도곡선 추가
              space = 0.2,          # x축, y축 공간 0으로 만들기
              ratio = 4           # 비율 조정하기
              )
```
