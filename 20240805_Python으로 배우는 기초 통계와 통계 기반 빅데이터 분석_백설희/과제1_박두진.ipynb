```python
# chunksize=1000 - 1000개씩 불러오기
chunker = pd.read_csv(os.path.join(rootdir, 'data', 'ex6.csv'), chunksize=1000)
```

```python
total_one = pd.Series([])
for piece in chunker:
  total_one = total_one.add(piece['one'].value_counts(), fill_value=0.0)
```

```python
total_one.head()
```

```python
print(total_one.sum())        # 전체 개수
print(total_one.count())      # 열 개수
```

```python
total_one_reset = total_one.reset_index()
total_one_reset
```

```python
# 합계 열 (sum) 추가
total_one_reset['sum'] = total_one_reset['one'] * total_one_reset[0]
```
