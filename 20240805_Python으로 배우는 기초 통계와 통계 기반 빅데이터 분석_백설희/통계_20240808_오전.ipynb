```python
import pandas as pd
import seaborn as sns

dia = sns.load_dataset('diamonds')
```

# 고차원함수(고계함수)
  - 함수의 입력매개변수로 함수를 받을 수 있는 특수한 함수

```python
dia.head()
```

```python
dia.info()
```

```python
dia2 = dia.iloc[:5]
dia2.shape
```

```python
dia2
```

## map 함수

### lambda

```python
dia2['carat_10'] = dia2['carat'].map(lambda x : x * 10)
```

```python
dia2
```

### 사용자 정의함수

```python
def myFunc1(x):
  return x * 10.0
```

```python
dia2['carat'].map(myFunc1)
```

```python
dia2[['carat', 'price']].map(myFunc1)
```

## apply 함수
- 열 단위 또는 행 단위로 함수를 적용

```python
def myFunc2(x):
  return x['carat'] * 10.0
```

```python
dia2.apply(myFunc2, axis = 1)
```

### z-score 표준화

```python
def z_std(x):
  return ((x - x.mean()) / x.std())
```

```python
dia2[['carat', 'price']].apply(z_std)
```

```python
dia2[['carat', 'price']].apply(z_std).describe()
```

### Min-Max Scale

```python
def min_max_scale(x):
  return (x - x.min()) / (x.max() - x.min())
```

```python
dia2[['carat', 'price']].apply(min_max_scale)
```

```python
dia2[['carat', 'price']].apply(min_max_scale).describe()
```

## 집계함수를 적용하는 고차원함수

### agg 함수
- 집계함수는 차원을 축소해준다 (1차원 -> 0차원)

```python
dia2.select_dtypes(include = 'number').columns
```

```python
dia2[dia2.select_dtypes(include = 'number').columns].agg(['sum', 'mean'])
```

#### 집계함수 사용자 정의

```python
def myMean(x):
  return x.sum() / x.count()
```

```python
dia2[dia2.select_dtypes(include = 'number').columns].agg([myMean])
```

```python
dia2[dia2.select_dtypes(include = 'number').columns].agg([z_std])
```
