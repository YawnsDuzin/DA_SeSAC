##continue는 반복문의 처음으로 다시 돌아간다

```python
txt = 'i am a boy'

for c in txt:
  if c == ' ':
    continue
  else:
    print('else')
  print('for end')
print('end')
```

with문을 이용한 파일 입출력
- 단순히 open()함수로 파일을 열었다면 작업이 끝난 후, close해 주어야 작업이 마무리됩니다. with문을 이용하면 이런 번거로움을 간단히 해결할 수 있습니다. with문은 파일을 열고 작업 이 끝난 후, close하지 않아도 자동으로 파일을 닫아 줍니다.
- close 하는 것을 자주 잊어 버린다면 차라리 with문을 이용해서 파일을 open하는 것이 안전합니다.

##Case2 - val개수를 동적으로 구성

```python
# 여러 행을 한까번에 읽어오기 (readlines())
fp = open(rootdir + '/data2.txt', 'r', encoding='utf-8')
lines = fp.readlines()      # 전체를 라인 단위로 리스트로 변환하여 반환

fp.close()
```

```python
dic1 = {}
list1 = []

# 데이터가 많은 경우에는 lines로 받으면 pc의 메모리 문제가 발생할 수 있다!
for line in lines:
  list1.append(line.replace('\n', '').split(' '))

print(list1)

for i in range(len(list1[0])):
  key = list1[0][i]

  # 동적으로 구성!
  val = []
  for j in range(1, len(list1)):
      val.append(list1[j][i])

  dic1[key] = val

print(dic1)
```

```python
fp = open(rootdir + '/data2.txt', 'r', encoding='utf-8')

head  = {}
data = {}

first = True

while True:
  line = fp.readline().replace('\n', '')
  if line == "" or line == "\n":
    break

  print('line : ', line)

  for i, item in enumerate(line.split(' ')):
    """
    0: 'Name'
    1: 'Kor'
    2: 'Eng'
    3: 'Math'
    4: 'His'
    """
    if first:
      head[i] = item
      data[item] = []
    else:
      """
      0: 'Lee'
      1: '90'
      2: '60'
      3: 'Math'
      4: 'His'
      """
      data[head[i]].append(item)
  first = False

print(head)
fp.close()

print(data)
```
