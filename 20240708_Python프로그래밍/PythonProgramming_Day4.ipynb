```python
# Case1
def not_used_char(txt):
  txts_lower = txt.lower()
  alpas = 'abcdefghijklmnopqrstuvwxyz'

  lst = []

  for alpa in alpas:
    if alpa not in txts_lower:
    # if txts_lower.find(alpa) == -1:
      lst.append(alpa)
  return lst
```

##함수구현

```python
def euclidean_dist(x1, y1, x2, y2):
    return ((x1-x2)**2 + (y1-y2)**2)**0.5
```

```python
def close_point(pts):
  clsd = []
  lst = list(pts.keys())    # 키 값 리스트로 변환
  lst_min= ['', '', 100]

  # 각 각의 두 점 사이의 거리 계산
  for i in range(len(lst)):
    for j in range(i+1, len(lst)):
      clsd.append([ lst[i], lst[j], euclidean_dist(pts[lst[i]][0], pts[lst[i]][1], pts[lst[j]][0], pts[lst[j]][1])])
      print('distance between {} and {}: {}'.format(clsd[-1][0], clsd[-1][1], clsd[-1][2]))

      if lst_min[2] > clsd[-1][2]:
        lst_min = clsd[-1]

  return lst_min
```

```python
print(close_point(pts))
```

```python
def close_point(pts):
  distns = {}

  for k1, v1 in pts.items():    # a ~ d
    for k2, v2 in pts.items():  # a ~ d
      # 동일한 거리는 제외할려고 추가
      # a-a, b-b, c-c는 계산할 필요가 없음!
      if k1 != k2:
        # 딕셔너리의 키는 리스트는 안됨 (튜플이나 문자열만 가능)
        # Ex) a-b계산, b-a는 계산이 필요없어서 제외하려고 추가!
        if distns.get(tuple(sorted([k1, k2]))) == None:
          distns[k1, k2] = ((v1[0]-v2[0])**2 + (v1[1]-v2[1])**2)**0.5

  print(distns)
  # 거리가 가장 가까운 것 리턴!!
  for k in distns.keys():
    if distns[k] == min(distns.values()):
      return k
```

```python
print(close_point(pts))
```

#금일 마지막 실습

```python
txt1 = 'I am a boy'
txt2 = 'You are a girl'

# 전체 소문자로 변환
# txt2 의 캐릭터가 txt1에 있으면,, txt2의 캐릭터를 *(아스테릭)으로 변경
```

##Case1. 쌩짜 구현

```python
def char_change(txt1, txt2):
  txt1 = txt1.lower()
  txt2 = txt2.lower()

  for c in txt1:
    txt2 = txt2.replace(c, '*')

  return txt2
```

```python
print(char_change(txt1, txt2))
```

##Case2. 비교문자 txt1의 중복 캐릭터 제거

```python
def char_change(txt1, txt2):
  lst1 = list(set(txt1.lower()))    # 중복제거(set변경후,list로 변경)

  for c in lst1:
    txt2 = txt2.lower().replace(c, '*')

  return txt2
```

```python
print(char_change(txt1, txt2))
```

##Case3. 정규식 사용

```python
import re

def char_change(txt1, txt2):
  return re.sub('['+txt1.lower()+']', '*', txt2.lower())
```

```python
print(char_change(txt1, txt2))
```

## Case4. 리스트 컴프리헨션

```python
def char_change(txt1, txt2):
  return ''.join([c if c not in list(set(txt1.lower())) else '*' for c in list(txt2.lower())])
```

```python
print(char_change(txt1, txt2))
```
