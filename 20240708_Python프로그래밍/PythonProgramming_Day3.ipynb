```python
def isSame(s1, s2):
  # 문자열 2개의 길이가 다른 경우 체크
  if len(s1) != len(s2):
    return False

  for i in range(len(s1)):
    if s1[i] != s2[i]:
      return False

  return True
```

```python
print(isSame(s1, s2))
print(isSame(s2, s3))
print(isSame('', ''))
print(isSame('', ' '))
```

###실급과제-find구현

```python
def myFind(s1, s2):
  for i in range(len(s1) - len(s2) + 1):   # 20 - 2 + 1 = 19 (0 ~ 18)
    if s1[i:(i + len(s2))] == s2:         # (0 ~ 18) : (2 ~ 20)
      return i

  return -1
```

```python
print(myFind('나는 학생입니다. 당신은 학생입니까?', '학생'))
print(myFind('나는 학생입니다. 당신은 학생입니까?', '당신은'))
print(myFind('나는 학생입니다. 당신은 학생입니까?', '니까?'))
print(myFind('나는 학생입니다. 당신은 학생입니까?', '군인'))
```

```python
# 여러 개 있는 경우, 다 찾기!
def myFind2(s1, s2):
  rtn = []

  for i in range(len(s1) - len(s2) + 1):   # 20 - 2 + 1 = 19 (0 ~ 18)
    if s1[i:(i + len(s2))] == s2:         # (0 ~ 18) : (2 ~ 20)
      rtn.append(i)

  return rtn
```

```python
def mySplit(s1):
  ret = []
  idx_start = 0

  for i in range(len(s1)):
    if s1[i] == ' ':
      ret.append(s1[idx_start:i])
      idx_start = i + 1

  # 마지막 공백 ~ 끝까지 확인
  ret.append(s1[idx_start:])

  return ret
```

```python
s1 = '나는 학생입니다. 당신은 학생입니까? '
print(mySplit(s1))
```

```python
def mySplit1(s1):
  ret = []
  idx_start = 0

  for i in range(len(s1)):
    if s1[i] == ' ':
      ret.append(s1[idx_start:i])
      idx_start = i + 1

      # 공백 후, 끝 부분 체크!
      if s1.find(' ', i + 1) == -1:
        ret.append(s1[idx_start:])
        break

  return ret
```

```python
s1 = '나는 학생입니다. 당신은 학생입니까?'
print(mySplit1(s1))
```

```python
def mySplit2(s1):
  ret = []
  spaces = []

  spaces = myFind2(s1, ' ')
  #print(spaces)
  #print(len(spaces))

  for i in range(len(spaces)):
    #print(i)
    # 시작일때는 앞쪽을 입력한다.
    if i == 0:
      ret.append(s1[:spaces[i]])
    # 끝 일때는 앞/뒤를 다 입력한다.
    elif i == (len(spaces) - 1):
      ret.append(s1[spaces[i-1]:spaces[i]])
      ret.append(s1[spaces[i]:])
    # 아닐때는 뒤쪽을 입력한다.
    else:
      ret.append(s1[spaces[i-1]:spaces[i]])

  return ret
```

```python
s1 = '나는 학생입니다. 당신은 학생입니까? '
print(mySplit2(s1))
```

```python
def mySplit3(s1):
  ret = []
  word = ''

  for c in s1:
    if c == ' ':
      ret.append(word)
      word = ''
    else:
      word += c

  # (공백~끝)마지막 부분 입력
  ret.append(word)

  return ret
```

###실습과제-item_count구현

```python
test_lst = [['a', 'b', 'c'], ['a', 'e', 'b'], ['a', 'c']]
print(test_lst)

def item_count(lst):
  ret1 = [] # 구분자 담을 변수
  ret2 = [] # 구분자 별 개수 담을 변수

  ret = []  # 마지막 리턴 변수

  for item in lst:
    for item2 in item:
      if ret1.count(item2) == 0:
         ret1.append(item2)
         ret2.append(1)
      else:
        ret2[ret1.index(item2)] += 1

  for i in range(len(ret1)):
    ret.append([ret1[i], ret2[i]])

  return ret
```

```python
print(item_count(test_lst))
```
