###Figure & AxesSubplot 객체

###눈금,라벨,범례

````python
fig = plt.figure(figsize=(5,2))
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())

# set_xticks() 함수로 x축 눈금 설정
# set_xtickLabels() 함수로 x축 눈금의 라벨 설정
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labeld = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=15, fontsize='small')

# set_title() 함수로 제목 설정
# set_xLabel() 함수로 x축 이름 설정
ax.set_title('My First matplotlib plot')
ax.set_xlabel('Stages')

#fig.subplots_adjust(bottom=0.3)

plt.show()
````

###범례 추가하기
- 각 그래프에 label을 지정한 후, ax.legend() 또는 plt.legend()를 실행하면 자동으로 범례가 생성된다.

````python
from numpy.random import randn

fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'g', label='one')
ax.plot(randn(1000).cumsum(), 'b--', label='two')
ax.plot(randn(1000).cumsum(), 'r.', label='three')

ax.legend(loc = 'best')     # 범례 표시

plt.show()
````

###막대 그래프
- 수직 막대 그래프를 그리려면 kind='bar', 수평 막대 그래프를 그리려면 kind='barh' 로 지정한다. 이 경우 Series 나 DataFrame의 색인은 X(bar) 눈금이나 Y(bar) 눈금을 사용한다.

````python
fig, axes = plt.subplots(2, 1, figsize=(6,5))
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot(kind='bar', ax=axes[0], color='y', alpha=0.7)
data.plot(kind='barh', ax=axes[1], color='y', alpha=0.7)
````

### 1. 전체 득표수에서 Bush, Gore, Etc 의 비율 (pie chart)

````python
# Bush, Gore Etc 득표수
Bush_vote = df1['Bush'].sum()                     # 2911215
Gore_vote = df1['Gore'].sum()                     # 2911417
Etc_vote = df1.iloc[:,6:].sum().sum()             # 137988

vals = [Bush_vote, Gore_vote, Etc_vote]
lbls = ['Bush', 'Gore', 'Etc']
cols = ["#dde5b6", "#adc178", "#a98467"]
#print(vals, lbls)
plt.pie(vals, labels=lbls, colors = cols, autopct='%.3f%%')
plt.legend(lbls)
#plt.title('전체 득표수에서 Bush, Gore, Etc 의 비율')
plt.show()
````

### 2. Bush, Gore 의 득표수와, 이긴 지역의 수 비교 (bar chart)

> 인용구 추가



````python
# Bush, Gore 득표수
Bush_vote = df1['Bush'].sum()                     # 2911215
Gore_vote = df1['Gore'].sum()                     # 2911417

# Bush, Gore 의 우승한 county 수
Bush_win = (df1['Bush'] > df1['Gore']).sum()      # 51
Gore_win = (df1['Bush'] < df1['Gore']).sum()      # 16

cols = ["#adc178", "#a98467"]

plt.subplot(1,2,1)
plt.bar(['Bush', 'Gore'], [Bush_vote, Gore_vote], color = cols)
plt.title('All votes')

plt.subplot(1,2,2)
plt.bar(['Bush', 'Gore'], [Bush_win, Gore_win], color = cols)
plt.title('Win Count')

plt.tight_layout()
plt.show()
````

### 3. Bush, Gore 의 각각 이긴 지역 중 득표차가 가장 많게 이긴 지역 5개 (bar chart)

````python
Bush_win_vote_term = df1.sort_values(by='Bush_win_vote_term', ascending=False)[:5][['Bush', 'Gore', 'Bush_win_vote_term']]
Gore_win_vote_term = df1.sort_values(by='Gore_win_vote_term', ascending=False)[:5][['Bush', 'Gore', 'Gore_win_vote_term']]

# print(Gore_win_vote_term['Gore_win_vote_term'])
# print(Gore_win_vote_term.index)

plt.subplot(1,2,1)
vals = Bush_win_vote_term['Bush_win_vote_term']
lbls = Bush_win_vote_term.index
plt.bar(lbls, vals)
plt.xticks(rotation=45)
plt.title('Bush_win')

plt.subplot(1,2,2)
vals = Gore_win_vote_term['Gore_win_vote_term']
lbls = Gore_win_vote_term.index
plt.bar(lbls, vals)
plt.xticks(rotation=45)
plt.title('Gore_win')

plt.tight_layout()
plt.show()
````

### 4. Bush, Gore 의 각각 이긴 지역 중 득표차가 가장 적게 이긴 지역 5개 (bar chart)
