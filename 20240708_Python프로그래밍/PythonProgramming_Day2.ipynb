```python
def my_in(num, lst):
  for i in range(len(lst)):
    if num == lst[i]:
      return True

  return False


print(my_in(5, [1,2,3,4,5]))
print(my_in(0, [1,2,3,4,5]))
```

```python
def my_in(num, lst):
  for i in lst:
    if num == i:
      return True

  return False


print(my_in(5, [1,2,3,4,5]))
print(my_in(0, [1,2,3,4,5]))
```

###Case2

```python
scores = [4.5, 3.0, 2.1, 3.7]
# enumerate : 인덱스, 값을 같이 반환
for _user, _score in enumerate(scores):
  msg3 = '%d 번 학생의 학점은 %1.1f 입니다.' % (_user + 1, _score)

  print(msg3)
```

```python
scores = [[2, 4.5], [5, 3.0], [9, 2.1], [21, 3.7]]
nd = ['첫', '두', '세', '네']
iCnt = 0
for _no, _score in scores:
  msg3 = '%2d 번 학생의 %s번째 학점은 %1.1f 입니다.' % (_no, nd[iCnt], _score)
  print(msg3)

  iCnt += 1
```
