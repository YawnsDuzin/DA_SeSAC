````python
a = input('비가 오는지 입력해주세요 (Y, N) : ')
b = input('우산이 있는지 입력해주세요 (Y, N) : ')
rain = True if a == 'Y' else False
umb = True if b == 'Y' else False
waitCnt = int(input('비오는데 우산없을때 기다릴 시간 : '))

def my_wait(icnt):
  print(str(icnt) + '-기다려')

def rain_check(rain, umb, waitCnt):
  icnt = 0

  if rain == True:
    print('비 온다')
    if umb == True:
      print('우산 있다.')
      print('외출 한다.')
    else:
      print('우산 없다.')
      while(rain == True):
        # waitCnt번 반복하고, 비안오게
        icnt = icnt + 1
        if icnt >= waitCnt:
          rain = False

        my_wait(icnt)

      print('우산 없다 - 비 안온다 - 외출 한다')
  else:
    print('비 안온다')
    print('외출 한다')

rain_check(rain, umb, waitCnt)
````

````python
RF_CARDS = ['VEHI_1', 'VEHI_2', 'VEHI_3', 'VEHI_4', 'VEHI_5']   # 저장된 차량카드번호

VEHI_WEIGHTS = []  # 저장된 차량무게

for i in range(1, 6):
  VEHI_WEIGHTS.append(i * 111)

INPUT_RF_CARD = input('차량카드(RF_CARD)를 입력해주세요 : ')
INPUT_VEHI_WEIGHT = int(input('현재 차량무게(VEHI_WEIGHT)를 입력해주세요 : '))

icnt = 0

def parsing_print(idx):
  msg = ['정상적으로 차량입문 하였습니다'   # 0
  , '차량카드가 입력되지 않았습니다.'       # 1
  , '차량무게가 입력되지 않았습니다.'       # 2
  , '입력/저장 차량무게가 같지 않습니다.'   # 3
  , '차량무게 입력 기다리는 중...']         # 4

  print(msg[idx])

def Vehi_CheckDetail(INPUT_VEHI_WEIGHT, VEHI_WEIGHT):
  return INPUT_VEHI_WEIGHT == VEHI_WEIGHT

def vehi_check(INPUT_RF_CARD, INPUT_VEHI_WEIGHT, VEHI_WEIGHT_CHANGE):
  global icnt
  icnt = 0
  chk = False

  if INPUT_RF_CARD.strip() != '':
    INDEX_RF_CARDS = RF_CARDS.index(INPUT_RF_CARD)

    if INPUT_VEHI_WEIGHT > 100:
      if Vehi_CheckDetail(INPUT_VEHI_WEIGHT, VEHI_WEIGHTS[INDEX_RF_CARDS]):
        parsing_print(0)
      else:
        parsing_print(3)
    else:
      while(chk == False):
        # 10번 반복
        icnt = icnt + 1
        if icnt >= 10:
          break;

        # VEHI_WEIGHT_CHANGE == True 면 차량입문 무게를 정상으로 처리 해준다!!
        if VEHI_WEIGHT_CHANGE == True and icnt == 5:
          INPUT_VEHI_WEIGHT = VEHI_WEIGHTS[INDEX_RF_CARDS]

        parsing_print(4)

        if INPUT_VEHI_WEIGHT == VEHI_WEIGHTS[INDEX_RF_CARDS]:
          chk = True;
          parsing_print(0)
  else:
    parsing_print(1)


print('시작')
vehi_check(INPUT_RF_CARD, INPUT_VEHI_WEIGHT, True)
print('종료')
````

###실습과제-find 함수구현

````python
def myFind(s1, s2):
  for i in range(len(s1) - len(s2) + 1):   # 20 - 2 + 1 = 19 (0 ~ 18)
    if s1[i:(i + len(s2))] == s2:         # (0 ~ 18) : (2 ~ 20)
      return i

  return -1
````

````python
print(myFind('나는 학생입니다. 당신은 학생입니까?', '학생'))
print(myFind('나는 학생입니다. 당신은 학생입니까?', '당신은'))
print(myFind('나는 학생입니다. 당신은 학생입니까?', '니까?'))
print(myFind('나는 학생입니다. 당신은 학생입니까?', '군인'))
````

````python
# 여러 개 있는 경우, 다 찾기!
def myFind2(s1, s2):
  rtn = []

  for i in range(len(s1) - len(s2) + 1):   # 20 - 2 + 1 = 19 (0 ~ 18)
    if s1[i:(i + len(s2))] == s2:         # (0 ~ 18) : (2 ~ 20)
      rtn.append(i)

  return rtn
````

###실습과제-split 함수구현

````python
def mySplit1(s1):
  ret = []
  idx_start = 0

  for i in range(len(s1)):
    if s1[i] == ' ':
      ret.append(s1[idx_start:i])
      idx_start = i + 1

      # 공백 후, 끝 부분 체크!
      if s1.find(' ', i + 1) == -1:
        ret.append(s1[idx_start:])
        break

  return ret
````

````python
s1 = '나는 학생입니다. 당신은 학생입니까?'
print(mySplit1(s1))
````

````python
def mySplit(s1):
  ret = []
  idx_start = 0

  for i in range(len(s1)):
    if s1[i] == ' ':
      ret.append(s1[idx_start:i])
      idx_start = i + 1

  # 마지막 공백 ~ 끝까지 확인
  ret.append(s1[idx_start:])

  return ret
````

````python
s1 = '나는 학생입니다. 당신은 학생입니까? '
print(mySplit(s1))
````

````python
def mySplit2(s1):
  ret = []
  spaces = []

  spaces = myFind2(s1, ' ')
  #print(spaces)
  #print(len(spaces))

  for i in range(len(spaces)):
    #print(i)
    # 시작일때는 앞쪽을 입력한다.
    if i == 0:
      ret.append(s1[:spaces[i]])
    # 끝 일때는 앞/뒤를 다 입력한다.
    elif i == (len(spaces) - 1):
      ret.append(s1[spaces[i-1]:spaces[i]])
      ret.append(s1[spaces[i]:])
    # 아닐때는 뒤쪽을 입력한다.
    else:
      ret.append(s1[spaces[i-1]:spaces[i]])

  return ret
````

````python
s1 = '나는 학생입니다. 당신은 학생입니까? '
print(mySplit2(s1))
````

````python
def mySplit3(s1):
  ret = []
  word = ''

  for c in s1:
    if c == ' ':
      ret.append(word)
      word = ''
    else:
      word += c

  # (공백~끝)마지막 부분 입력
  ret.append(word)

  return ret
````

###실습과제-item_count구현
- 문자별 빈도 수 계산하여 리턴하는 함수

````python
test_lst = [['a', 'b', 'c'], ['a', 'e', 'b'], ['a', 'c']]
print(test_lst)

def item_count(lst):
  ret1 = [] # 구분자 담을 변수
  ret2 = [] # 구분자 별 개수 담을 변수

  ret = []  # 마지막 리턴 변수

  for item in lst:
    for item2 in item:
      if ret1.count(item2) == 0:
         ret1.append(item2)
         ret2.append(1)
      else:
        ret2[ret1.index(item2)] += 1

  for i in range(len(ret1)):
    ret.append([ret1[i], ret2[i]])

  return ret
````

###함수구현(유클리드 거리)

````python
def euclidean_dist(x1, y1, x2, y2):
    return ((x1-x2)**2 + (y1-y2)**2)**0.5
````

````python
def close_point(pts):
  clsd = []
  lst = list(pts.keys())    # 키 값 리스트로 변환
  lst_min= ['', '', 100]

  # 각 각의 두 점 사이의 거리 계산
  for i in range(len(lst)):
    for j in range(i+1, len(lst)):
      clsd.append([ lst[i], lst[j], euclidean_dist(pts[lst[i]][0], pts[lst[i]][1], pts[lst[j]][0], pts[lst[j]][1])])
      print('distance between {} and {}: {}'.format(clsd[-1][0], clsd[-1][1], clsd[-1][2]))

      if lst_min[2] > clsd[-1][2]:
        lst_min = clsd[-1]

  return lst_min
````

````python
print(close_point(pts))
````

````python
def close_point(pts):
  distns = {}

  for k1, v1 in pts.items():    # a ~ d
    for k2, v2 in pts.items():  # a ~ d
      # 동일한 거리는 제외할려고 추가
      # a-a, b-b, c-c는 계산할 필요가 없음!
      if k1 != k2:
        # 딕셔너리의 키는 리스트는 안됨 (튜플이나 문자열만 가능)
        # Ex) a-b계산, b-a는 계산이 필요없어서 제외하려고 추가!
        if distns.get(tuple(sorted([k1, k2]))) == None:
          distns[k1, k2] = ((v1[0]-v2[0])**2 + (v1[1]-v2[1])**2)**0.5

  print(distns)
  # 거리가 가장 가까운 것 리턴!!
  for k in distns.keys():
    if distns[k] == min(distns.values()):
      return k
````

````python
def char_change(txt1, txt2):
  txt1 = txt1.lower()
  txt2 = txt2.lower()

  for c in txt1:
    txt2 = txt2.replace(c, '*')

  return txt2
````

###Case2 - val개수를 동적으로 구성

````python
# 여러 행을 한까번에 읽어오기 (readlines())
fp = open(rootdir + '/data2.txt', 'r', encoding='utf-8')
lines = fp.readlines()      # 전체를 라인 단위로 리스트로 변환하여 반환

fp.close()
````

````python
dic1 = {}
list1 = []

# 데이터가 많은 경우에는 lines로 받으면 pc의 메모리 문제가 발생할 수 있다!
for line in lines:
  list1.append(line.replace('\n', '').split(' '))

print(list1)

for i in range(len(list1[0])):
  key = list1[0][i]

  # 동적으로 구성!
  val = []
  for j in range(1, len(list1)):
      val.append(list1[j][i])

  dic1[key] = val

print(dic1)
````

###Case3 - 교수님

|Name|Kor|Eng|Math|His|
|:---:|:---|:---|:---|:---|
|Lee|90|50|70|11|
|Kim|50|60|90|22|
|Han|90|95|100|33|
|Park|85|90|95|44|

````python
fp = open(rootdir + '/data2.txt', 'r', encoding='utf-8')

head  = {}
data = {}

first = True

while True:
  line = fp.readline().replace('\n', '')
  if line == "" or line == "\n":
    break

  print('line : ', line)

  for i, item in enumerate(line.split(' ')):
    """
    0: 'Name'
    1: 'Kor'
    2: 'Eng'
    3: 'Math'
    4: 'His'
    """
    if first:
      head[i] = item
      data[item] = []
    else:
      """
      0: 'Lee'
      1: '90'
      2: '60'
      3: 'Math'
      4: 'His'
      """
      data[head[i]].append(item)
  first = False

print(head)
fp.close()

print(data)
````
