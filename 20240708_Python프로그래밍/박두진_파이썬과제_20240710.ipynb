```python
test_lst = [['a', 'b', 'c'], ['a', 'e', 'b'], ['a', 'c']]
print(test_lst)
```

```python
def item_count(lst):
  ret1 = [] # 구분자 담을 변수
  ret2 = [] # 구분자 별 개수 담을 변수

  ret = []  # 마지막 리턴 변수

  for item in lst:
    for item2 in item:
      if ret1.count(item2) == 0:
         ret1.append(item2)
         ret2.append(1)
      else:
        ret2[ret1.index(item2)] += 1

  for i in range(len(ret1)):
    ret.append([ret1[i], ret2[i]])

  return ret
```

```python
print(item_count(test_lst))
```
